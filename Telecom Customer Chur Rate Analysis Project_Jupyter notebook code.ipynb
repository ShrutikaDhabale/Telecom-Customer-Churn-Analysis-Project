##Outlier Treatment

```python
for i in data.columns:
  if data[i].dtype!='object':
    sns.boxplot(data[i])
    plt.show()
```

Above observation will showing most of the integer fatures having outlier so data need outlier treatment.

```python

for i in data.columns:
  if data[i].dtype!='object':
    Q1 = data[i].quantile(0.25)
    Q3 = data[i].quantile(0.75)
    IQR = Q3 - Q1
    Outer_layer = Q3+1.5*IQR
    Lower_layer = Q1-1.5*IQR
    Outer_layer,Lower_layer

    data.loc[data[i] <Lower_layer, i] = Lower_layer
    data.loc[data[i] >Outer_layer, i] = Outer_layer
```

```python
space={ 
        #'booster': hp.choice('booster',['gbtree', 'dart']),
        'max_depth': hp.quniform("max_depth", 2, 22, 1),
        #'gamma': hp.uniform ('gamma', 0, 1),
        'reg_alpha' : hp.quniform('reg_alpha', 1,50,1),
        'reg_lambda' : hp.uniform('reg_lambda', 0,1),
        'colsample_bytree' : hp.uniform('colsample_bytree', 0.6, 1),
        #'min_child_weight' : hp.quniform('min_child_weight', 1, 40,1),
        'n_estimators': hp.quniform('n_estimators', 1000,3500,100),
        'scale_pos_weight':hp.quniform("scale_pos_weight", 1, 5, 1),       
        'learning_rate': hp.uniform('learning_rate', 0.01, 0.1),
        #'subsample': hp.uniform('subsample', 0.5, 0.9),
        'colsample_bylevel' : hp.uniform('colsample_bylevel', 0.6, 1),
        #'max_delta_step' : hp.quniform('max_delta_step', 1, 10,1),
        #'eta' : hp.uniform ('eta', 0, 1),
        'seed': 0
    }
```

```python
def objective(space):
    clf=XGBClassifier(
                    #objective = 'reg:logistic',
                    #booster = str(space['booster']),
                    max_depth = int(space['max_depth']),
                    #gamma = space['gamma'],
                    reg_alpha = int(space['reg_alpha']),
                    reg_lambda = space['reg_lambda'],
                    colsample_bytree = space['colsample_bytree'],
                    #min_child_weight = int(space['min_child_weight']),
                    n_estimators = int(space['n_estimators']),
                    scale_pos_weight = int(space['scale_pos_weight']),
                    learning_rate = space['learning_rate'],
                    #subsample = space['subsample'],
                    colsample_bylevel = space['colsample_bylevel'],
                    #max_delta_step = int(space['max_delta_step']),
                    #eta = space['eta'],
                    seed = 0
                    )
    
    evaluation = [(x_train1, y_train), (x_test1, y_test)]
    
    clf.fit(x_train1, y_train,
            eval_set=evaluation, eval_metric="auc",
            early_stopping_rounds=10,verbose=False)
    
    score = metrics.recall_score(y_test, clf.predict(x_test1))
    #score = metrics.accuracy_score(y_val, clf.predict(X_val))

    print ("SCORE:", score)
    return {'loss': -score, 'status': STATUS_OK }
```

```python
import sklearn.metrics as metrics
trials = Trials()
best_hyperparams = fmin(fn = objective,
                        space = space,
                        algo = tpe.suggest,
                        max_evals = 10,
                        trials = trials)
```
